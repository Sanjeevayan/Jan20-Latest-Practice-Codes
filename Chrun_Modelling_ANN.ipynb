```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.preprocessing import MinMaxScaler,StandardScaler
```

```python
scalar=StandardScaler()
X_train,X_test,y_train,Y_test=train_test_split(X2,y,test_size=0.2)
```

```python
X_train = scalar.fit_transform(X_train)
```

```python
X_test=scalar.transform(X_test)
```

```python
X_train,X_test
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import ReLU,LeakyReLU,PReLU,ELU
from keras.layers import Dropout
```

```python
model = Sequential()
model.add(Dense(output_dim=6, init="he_uniform",activation="relu",input_dim=10))

#classifer.add(Dense(output_dim = 6, init ='he_uniform',activation='relu',input_dim = 11))
#output_dim = 6, init = 'he_uniform',activation='relu',input_dim = 11)
```

```python
from keras.layers import LSTM
#Adding second hidden layer
# Adding the second hidden layer
model.add(Dense(output_dim = 6, init = 'he_uniform',activation='relu'))
#model.add(LSTM(output_dim = 6, init = 'he_uniform',activation='relu'))
```

```python
#adding Output layer
model.add(Dense(output_dim=1,init="he_uniform",activation="sigmoid"))
```

```python
#compiling the model
model.compile(optimizer = 'Adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
#model_history=model.fit(X_train,y_train,validation_split=0.33,batch_size=10,epochs=50)

model_history=model.fit(X_train, y_train,validation_split=0.33, batch_size = 10, nb_epoch = 100)
```

```python
import matplotlib.pyplot as plt
plt.plot(model_history.history['acc'])
plt.plot(model_history.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
 
# summarize history for loss
plt.plot(model_history.history['loss'])
plt.plot(model_history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
# Predicting the Test set results
y_pred = model.predict(X_test)
y_pred = (y_pred > 0.5)

# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, y_pred)

# Calculate the Accuracy
from sklearn.metrics import accuracy_score
score=accuracy_score(y_pred,Y_test)
cm,score
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(Y_test, y_pred)
```
